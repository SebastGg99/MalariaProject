```python
# Lattice y parmetros
L = LatticeSOS(size=10, seed=42)
L.initialize("random_surface")

params = KMCParams(
    T=298.15,
    K0_plus=1.9,
    K_inc_plus=1.9,
    E_pb_over_kT=0.7,
    phi_over_kT= 1.9,
    delta=6.9,
    V=1.0,
    C_eq=5.0,
    #C_eq=20.0, # conv_max ~ 95 %
    S_floor=-5.0, S_ceil=8.0
)

kmc = KMC_BKL(L, params, N_bulk0=3000, rng_seed=123, time_scale=1.0, n_seeds=70)


times = np.linspace(0, 100, 25)
snaps = kmc.run(t_end=1, snapshot_times=times)


#plt.plot(Experiments[2]['Tiempo'], Experiments[2]['Porcentaje'], label='Datos Experimentales')
plt.plot([t/2.5 for t,_,_ in snaps],
         [conv for _,_,conv in snaps], marker="o")
plt.xlabel("Tiempo simulado (h)")
plt.ylabel("Conversin (%)")
plt.title("BKL con adtomos explcitos (forma en S)")
plt.grid(True)
plt.show()
```
